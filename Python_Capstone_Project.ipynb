**Problem Statement:**

In the highly competitive landscape of digital marketing, effectively evaluating the success of
various marketing campaigns is essential for optimizing return on investment (ROI) and

#3) Exploratory Data Analysis (EDA) and Visualization

**Campaign Performance:**

âž¢ Plot a scatter plot to visualize the relationship between Acquisition_Cost and
ROI.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Acquisition_Cost', y='ROI')
plt.title('Acquisition Cost vs ROI')
plt.show()
```

âž¢ Create a bar chart to visualize the average Conversion_Rate for different
Channel_Used, categorized by Campaign_Type.

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Channel_Used', y='Conversion_Rate', hue='Campaign_Type')
plt.title('Average Conversion Rate by Channel and Campaign Type')
plt.show()
```

âž¢ Visualize the distribution of Engagement_Score across different Campaign_Type
using a box plot.

```python
sns.boxplot(data=df, x='Campaign_Type', y='Engagement_Score')
plt.title('Distribution of Engagement Score by Campaign Type')
plt.show()
```

âž¢ Analyze the average ROI by Company using a bar chart to compare the profitability
of campaigns conducted by different companies.

```python
ax = sns.barplot(data=df, x='Company', y='ROI' , estimator=pd.Series.mean ,ci=None)
ax.bar_label(ax.containers[0])
plt.title('Average ROI by Company')
plt.xticks(rotation=45)
plt.show()
```

âž¢ Examine the correlation between Engagement_Score and Conversion_Rate
using a heatmap.

```python
sns.heatmap(df[['Engagement_Score', 'Conversion_Rate']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**Customer Segmentation**

âž¢ Create a count plot to visualize the distribution of Target_Audience.


```python
ax = sns.countplot(data=df, x='Target_Audience')
ax.bar_label(ax.containers[0])
plt.title('Distribution of Target Audience')
plt.xticks(rotation=45)
plt.show()
```

âž¢ Identify which Customer_Segment has the highest Conversion_Rate for each
Language using a bar chart.


```python
sns.barplot(data=df, x='Customer_Segment', y='Conversion_Rate', hue='Language')
plt.title('Conversion Rate by Customer Segment and Language')
plt.xticks(rotation=45)
plt.show()
```

âž¢ Visualize the distribution of Acquisition_Cost across each Customer_Segment,
categorized by Channel_Used, using a box plot.


```python
sns.barplot(data=df, x='Customer_Segment', y='Acquisition_Cost', hue='Channel_Used')
plt.title('Distribution of Acquisition Cost by Customer Segment and Channel Used')
plt.xticks(rotation=45)
plt.show()
```

âž¢ Analyze average Conversion_Rate by Language using a bar chart to compare the effectiveness of campaigns conducted in different languages.

```python
sns.barplot(data=df, x='Language', y='Conversion_Rate', hue = 'Language' , ci=None)
plt.title('Average Conversion Rate by Language')
plt.xticks(rotation=45)
plt.show()
```

**Channel Effectiveness:**

âž¢ Compare the Engagement_Score for different Channels_Used, segmented by
Campaign_Type, using a bar chart.


```python
plt.figure(figsize=(12,6))
sns.barplot(data=df, x='Channel_Used', y='Engagement_Score', hue='Campaign_Type')
plt.title('Engagement Score by Channel and Campaign Type')
plt.show()
```

âž¢ Show the distribution of total ROI across different Channels_Used using a pie
chart.


```python
plt.pie(df.groupby('Channel_Used')['ROI'].sum(), labels=df['Channel_Used'].unique(), autopct='%1.1f%%')
plt.title('Distribution of Total ROI by Channel')
plt.show()
```

âž¢ Plot a scatter plot to show the relationship between Clicks and Impressions for each Channel_Used.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Clicks', y='Impressions', hue='Channel_Used')
plt.title('Relationship between Clicks and Impressions by Channel')
plt.show()
```

**Time-Based Analysis:**

âž¢ Plot the distribution of Duration using a histogram.


```python
ax = sns.histplot(data=df, x='Duration', bins=4 , kde=True)
ax.bar_label(ax.containers[0])
plt.title('Distribution of Duration')
plt.show()
```

âž¢ Analyze how the overall Conversion_Rate has changed over Date for each
Company using a line chart.


```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Date', y='Conversion_Rate', hue='Company')
plt.title('Conversion Rate Over Time by Company')
plt.show()
```

âž¢ Examine the trend of Engagement_Score over Date with a line chart.

```python
sns.lineplot(data=df, x='Date', y='Engagement_Score')
plt.title('Engagement Score Over Time')
plt.show()
```

**Geographic Analysis:**

âž¢ Determine which location has the highest Acquisition_Cost using a bar chart.


```python
plt.figure(figsize=(10,6))
ax = sns.barplot(data=df, x='Location', y='Acquisition_Cost', ci=None)
ax.bar_label(ax.containers[0])
plt.title('Acquisition Cost by Location')
plt.xticks(rotation=45)
plt.show()
```

âž¢ Visualize the Conversion_Rate by different Location, categorized by
Target_Audience, using a bar chart.


```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Location', y='Conversion_Rate', hue='Target_Audience')
plt.title('Conversion Rate by Location and Target Audience')
plt.xticks(rotation=45)
plt.show()
```

âž¢ Illustrate the proportion of ROI by Location using a pie chart.

```python
plt.pie(df.groupby('Location')['ROI'].sum(), labels=df['Location'].unique(), autopct='%1.1f%%')
plt.title('Proportion of Total ROI by Location')
plt.show()
```
